```python
# --- ベースのDataFrameをコピー ---
df_base = df_filtered.copy()

# === 🔹 15万kg以上の日付を除外 ===
daily_total = df_base.groupby("伝票日付")["正味重量"].sum()

# 13万kg以上の日付を抽出
high_volume_dates = daily_total[daily_total >= 130000].index.tolist()

# 除外
df_base = df_base[~df_base["伝票日付"].isin(high_volume_dates)].copy()

# === 🔸 下位5%日付（2万kg未満）の除外 ===

# ① 平日・土曜 & 非祝日 のフラグを追加
df_base["is_valid_day"] = (df_base["weekday"] != 6) & (~df_base["祝日フラグ"])

# ② 該当日の日別合計を算出
valid_daily_total = (
    df_base[df_base["is_valid_day"]]
    .groupby("伝票日付")["正味重量"]
    .sum()
)

# ③ 下位5%のしきい値を算出
lower_threshold = valid_daily_total.quantile(0.05)

# ④ 下位5%かつ2万kg未満の日付を抽出
low_volume_dates = valid_daily_total[
    (valid_daily_total < lower_threshold) & (valid_daily_total < 20000)
].index.tolist()

# ⑤ 除外対象を取り除いたdfを作成（mainフィルタ済みデータ）
df_filtered_final = df_base[~df_base["伝票日付"].isin(low_volume_dates)].copy()

# 結果確認用（任意）
df_filtered_final
```

## 将来予測モデル作成

```python
# === 必要なライブラリ ===
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

# --- 祝日フラグ判定関数 ---
def make_holiday_flag(date):
    import jpholiday
    return int(jpholiday.is_holiday(date) or date.weekday() >= 5)

# --- 品目分類関数 ---
def classify_item(item: str) -> str:
    if "混合廃棄物A" in item:
        return "混合A"
    elif "混合廃棄物B" in item:
        return "混合B"
    else:
        return "その他"

# --- データ準備 ---
df = df_pickup.copy()
df["伝票日付"] = pd.to_datetime(df["伝票日付"])
df["品目分類"] = df["品名"].apply(classify_item)

# --- 日次集計 ---
df_daily = df.groupby(["伝票日付", "品目分類"])["正味重量"].sum().reset_index()

# --- ピボット ---
df_pivot = df_daily.pivot(index="伝票日付", columns="品目分類", values="正味重量").fillna(0)

# --- 特徴量作成 ---
df_feat = df_pivot.copy()
df_feat["曜日"] = df_feat.index.dayofweek
df_feat["月"] = df_feat.index.month
df_feat["祝日フラグ"] = df_feat.index.map(make_holiday_flag)
df_feat["前日合計"] = df_pivot.sum(axis=1).shift(1)

# --- 前日_分類別特徴量 ---
for col in df_pivot.columns:
    df_feat[f"前日_{col}"] = df_pivot[col].shift(1)

# --- 目的変数（翌日の合計重量） ---
df_feat["翌日合計"] = df_pivot.sum(axis=1).shift(-1)

# --- 欠損除去 ---
df_feat = df_feat.dropna().reset_index()

# === 説明変数 / 目的変数 ===
X = df_feat.drop(columns=["伝票日付", "翌日合計"])
y = df_feat["翌日合計"]

# === データ分割 ===
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# === LightGBMデータセット作成 ===
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)

# === パラメータ設定 ===
params = {
    "objective": "regression",
    "metric": "mae",
    "verbosity": -1,
    "boosting_type": "gbdt",
}

# === 学習 ===
model = lgb.train(
    params,
    lgb_train,
    valid_sets=[lgb_train, lgb_eval],
    num_boost_round=1000,
    early_stopping_rounds=50,
    verbose_eval=100,
)

# === 予測 ===
y_pred = model.predict(X_test)

# === 評価 ===
mae = mean_absolute_error(y_test, y_pred)
print(f"\n📊 MAE（平均絶対誤差）: {mae:.2f} kg")

# === 予測 vs 実測 可視化 ===
plt.figure(figsize=(12, 6))
plt.plot(y_test.values, label="実測", marker="o")
plt.plot(y_pred, label="予測", marker="x")
plt.legend()
plt.title("LightGBMによる搬入量予測（全特徴量使用）")
plt.xlabel("テストデータの時間順インデックス")
plt.ylabel("翌日正味重量")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
from tqdm import tqdm

# 結果格納用
model_results = {}
y_preds = {}

# tqdm付きで品目ごとにモデル学習
for target_item in tqdm(df_pivot.columns, desc="📦 モデル学習中（品目別）"):

    y = df_pivot[target_item]
    X = df_feat[["曜日", "月", "祝日フラグ", "前日合計"]].copy()

    # 欠損除外
    mask = y.notna()
    X_sub, y_sub = X[mask], y[mask]

    # 学習データが極端に少ない品目はスキップ
    if len(y_sub) < 30:
        print(f"⚠️ {target_item} は学習データが少ないためスキップ")
        continue

    # データ分割
    X_train, X_test, y_train, y_test = train_test_split(
        X_sub, y_sub, test_size=0.2, shuffle=False
    )

    # モデル学習
    model = RandomForestRegressor(random_state=42)
    model.fit(X_train, y_train)

    # 予測
    y_pred = model.predict(X_test)

    # 評価
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    model_results[target_item] = {
        "model": model,
        "R²": r2,
        "MAE": mae,
    }
    y_preds[target_item] = (y_test, y_pred)

# 結果確認
for item, res in model_results.items():
    print(f"📘 {item}")
    print(f"　　R² (test) = {res['R²']:.3f}")
    print(f"　　MAE (test) = {int(res['MAE'])} kg\n")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_absolute_error
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# --- サンプルのdf_featとdf_pivotを作成（本来は前処理後のものを使用） ---
np.random.seed(42)
dates = pd.date_range(start="2024-01-01", periods=120, freq="D")
df_pivot = pd.DataFrame({
    "混合A": np.random.randint(5000, 10000, size=120),
    "混合B": np.random.randint(2000, 6000, size=120),
    "その他": np.random.randint(1000, 7000, size=120)
}, index=dates)

df_feat = df_pivot.copy()
df_feat["曜日"] = df_feat.index.dayofweek
df_feat["月"] = df_feat.index.month
df_feat["祝日フラグ"] = df_feat["曜日"].isin([5, 6]).astype(int)
df_feat["前日合計"] = df_pivot.sum(axis=1).shift(1)

# 前日品目別を追加
for col in df_pivot.columns:
    df_feat[f"{col}_前日"] = df_pivot[col].shift(1)

# --- ステージ1: LightGBMで品目別モデルを学習 ---
item_models = {}
for item in df_pivot.columns:
    y = df_pivot[item]
    feature_cols = ["曜日", "月", "祝日フラグ", "前日合計"] + [f"{col}_前日" for col in df_pivot.columns]
    X = df_feat[feature_cols].copy()
    mask = y.notna()
    X_sub, y_sub = X[mask], y[mask]
    if len(y_sub) < 30:
        continue
    X_train, X_test, y_train, y_test = train_test_split(X_sub, y_sub, test_size=0.2, shuffle=False)
    model = lgb.LGBMRegressor(random_state=42)
    model.fit(X_train, y_train)
    item_models[item] = model

# --- ステージ2: 各日の予測値を特徴量としてメタモデルを学習 ---
stage2_X = []
stage2_y = []

valid_dates = df_feat.index[10:]  # 最初の10日は前日データ不足回避
for day in valid_dates:
    if day not in df_feat.index:
        continue
    x_row = df_feat.loc[day].copy()
    preds = []
    for item in df_pivot.columns:
        model = item_models.get(item)
        if model is None:
            preds.append(0)
            continue
        feature_input = x_row[["曜日", "月", "祝日フラグ", "前日合計"] + [f"{col}_前日" for col in df_pivot.columns]].to_frame().T
        pred = model.predict(feature_input)[0]
        preds.append(pred)
    # メタモデル用特徴量 + y
    stage2_X.append(preds + [x_row["曜日"], x_row["月"], x_row["祝日フラグ"], x_row["前日合計"]])
    stage2_y.append(df_pivot.loc[day].sum())

# データフレームに変換
columns = ["混合A予測", "混合B予測", "その他予測", "曜日", "月", "祝日フラグ", "前日合計"]
df_stage2_X = pd.DataFrame(stage2_X, columns=columns)
df_stage2_y = pd.Series(stage2_y)

# --- メタモデル学習 ---
meta_model = Ridge()
meta_model.fit(df_stage2_X, df_stage2_y)

# --- 評価 ---
y_pred = meta_model.predict(df_stage2_X)
r2 = r2_score(df_stage2_y, y_pred)
mae = mean_absolute_error(df_stage2_y, y_pred)


# 精度指標を出力
print(f"\n📊 メタモデル評価指標:")
print(f"R² = {r2:.3f}")
print(f"MAE = {int(mae)} kg")

# グラフ表示
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(df_stage2_y.values, label="実測値（合計）", marker='o')
plt.plot(y_pred, label="予測値（合計）", marker='x')
plt.title("全体搬入量（スタッキング予測）")
plt.xlabel("予測日数")
plt.ylabel("重量 [kg]")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("meta_model_prediction_plot.png")  # 画像保存
plt.show()
```

```python
import requests
import pandas as pd

def get_tokyo_weather_history(start_date: str, end_date: str) -> pd.DataFrame:
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude":35.645213,
        "longitude": 139.838695,
        "start_date": start_date,
        "end_date": end_date,
        "daily": ["temperature_2m_max", "temperature_2m_min", "precipitation_sum"],
        "timezone": "Asia/Tokyo"
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data["daily"])
        df["time"] = pd.to_datetime(df["time"])
        return df
    else:
        print("❌ Failed to fetch data:", response.text)
        return pd.DataFrame()
# 取得
df_weather = get_tokyo_weather_history("2020-01-01", "2025-06-01")
```

```python
def classify_weather(row):
    if row["precipitation_sum"] == 0.0:
        return "clear"
    elif row["precipitation_sum"] <= 1.0:
        return "light rain"
    elif row["precipitation_sum"] <= 10.0:
        return "rain"
    else:
        return "heavy rain"

df_weather["weather_label"] = df_weather.apply(classify_weather, axis=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import joblib

# === 特徴量生成 ===
df_selected["前日重量"] = df_selected["正味重量"].shift(1)
df_selected["前週平均"] = df_selected["正味重量"].shift(1).rolling(window=7).mean()
df_selected["前日比"] = df_selected["正味重量"] / df_selected["前日重量"]
df_selected = df_selected.replace([np.inf, -np.inf], np.nan).dropna()

# === 説明変数・目的変数 ===
X = df_selected[["weekday", "month", "祝日フラグ", "天気", "前日重量", "前週平均", "前日比"]].copy()
y = df_selected["正味重量"]

# === One-Hot Encoding（weekday, 天気） ===
X_encoded = pd.get_dummies(X, columns=["weekday", "天気"], drop_first=True)

# === データ分割 ===
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# === 特徴量選択 ===
base_model = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
base_model.fit(X_train, y_train)

selector = SelectFromModel(base_model, threshold="median")
selector.fit(X_train, y_train)
selected_features = X_encoded.columns[selector.get_support()]
print("✅ 選ばれた特徴量:", selected_features.tolist())

# === 特徴量重要度のプロット ===
importances = base_model.feature_importances_
features = X_encoded.columns
importance_df = pd.DataFrame({
    "特徴量": features,
    "重要度": importances
}).sort_values(by="重要度", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(importance_df["特徴量"], importance_df["重要度"])
plt.xlabel("重要度")
plt.title("特徴量の重要度（削減前）")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# === 特徴量選択後のデータに変換 ===
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

# === 再学習（本番用モデル） ===
model = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
model.fit(X_train_selected, y_train)

# === 評価（R²スコア） ===
y_train_pred = model.predict(X_train_selected)
y_test_pred = model.predict(X_test_selected)

print("🎯 R²（訓練データ）:", r2_score(y_train, y_train_pred))
print("🧪 R²（テストデータ）:", r2_score(y_test, y_test_pred))

# === 散布図：予測 vs 実測 ===
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_test_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("実測値")
plt.ylabel("予測値")
plt.title("予測値 vs 実測値（テストデータ）")
plt.tight_layout()
plt.show()

# === モデル保存 ===
joblib.dump(model, "rf_model_selected_v2.joblib")
print("✅ モデルを保存しました → rf_model_selected_v2.joblib")
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
import numpy as np
import pandas as pd

# === 特徴量生成（すでに作成済前提） ===
df_selected["前日重量"] = df_selected["正味重量"].shift(1)
df_selected["前週平均"] = df_selected["正味重量"].shift(1).rolling(window=7).mean()
df_selected["前日との差"] = df_selected["前日重量"] - df_selected["前週平均"]
df_selected = df_selected.replace([np.inf, -np.inf], np.nan).dropna()
df_selected = df_selected.sort_index()

X = df_selected[["weekday", "month", "祝日フラグ", "天気", "前日重量", "前週平均", "前日との差"]].copy()
y = df_selected["正味重量"]
X_encoded = pd.get_dummies(X, columns=["weekday", "天気"], drop_first=True)

# === TimeSeriesSplit による検証 ===
tscv = TimeSeriesSplit(n_splits=5)
r2_scores = []

for fold, (train_index, test_index) in enumerate(tscv.split(X_encoded)):
    X_train, X_test = X_encoded.iloc[train_index], X_encoded.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # 特徴量選択
    selector = SelectFromModel(RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42), threshold="median")
    selector.fit(X_train, y_train)
    X_train_sel = selector.transform(X_train)
    X_test_sel = selector.transform(X_test)

    # モデル学習
    model = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
    model.fit(X_train_sel, y_train)

    # 評価
    y_pred = model.predict(X_test_sel)
    score = r2_score(y_test, y_pred)
    print(f"📦 Fold {fold+1} R²: {score:.4f}")
    r2_scores.append(score)

print("\n📈 平均R²スコア（TimeSeriesSplit）:", np.round(np.mean(r2_scores), 4))
```
